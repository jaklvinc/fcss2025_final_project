## See what different main paths there are

```python
articles['MainPath'] = articles['Path'].apply(lambda x: x.split('/')[0])
```

```python
articles['MainPath'].value_counts()
```

```python
news = articles[articles['MainPath'] == 'Newsroom'].copy()
news['NewsroomTopic'] = news['Path'].apply(lambda x: x.split('/')[1])
news['NewsroomTopic'].value_counts()
```
